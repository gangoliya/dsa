## Bubble Sort

```python
def bubble_sort(nums):
    # Create a copy of the list, to avoid changing it
    nums = list(nums)

    # 4. Repeat the process n-1 time
    for _ in range(len(nums) - 1):

        # 1. Iterate over the array (except last element)
        for i in range(len(nums) - 1):

            # 2. Compare the number with
            if nums[i] > nums[i+1]:

                # 3. Swap the two elements
                nums[i], nums[i+1] = nums[i+1], nums[i]
    
    # Return the sorted list
    return nums
```

## Insertion Sort  


```python
def insertion_sort(nums):
    nums = list(nums) # made a copy of the list nums called nums
    for i in range(len(nums)):
        cur = nums.pop(i) # deleting the value at index i
        j = i-1
        while j >= 0 and nums[j] > cur:
            j -= 1
        nums.insert(j+1, cur)
    return nums
```

```python
def merge_sort(nums):
    # Terminating condition (list of 0 or 1 elements)
    if len(nums) <= 1:
        return nums
    
    # Get the midpoint
    mid = len(nums)//2

    # Split the list into two halves
    list1 = nums[:mid]
    list2 = nums[mid:]

    # Solve the problem for each half recursively
    left_sorted, right_sorted = merge_sort(list1), merge_sort(list2)

    # Combine the results of the two halves
    sorted_nums = merge(left_sorted, right_sorted)

    return sorted_nums
```

```python
def merge(nums1, nums2):
    # List to store teh results
    merged = []

    # Indices for iteration
    i = j = 0

    # Loop over the two list
    while i < len(nums1) and j < len(nums2):

        # Include the smaller element in the result and move to the next element
        if nums1[i] <= nums2[j]:
            merged.append(nums1[i])
            i += 1
        else:
            merged.append(nums2[j])
            j += 1

    # Get the remaining parts
    # At the end of the above while loop one of the two list will be exhausted and the code below will get the remaining parts from the loop with elements
    nums1_tail = nums1[i:]
    nums2_tail = nums2[j:]

    # Return the final merged array
    return merged + nums1_tail + nums2_tail
```

## 9. Analyze the algorithm's complexity and identify inefficiencies  

```python
def merge(nums1, nums2, depth=0):
    print(' '*depth, 'merge:', nums1, nums2)
    i, j, merged = 0, 0, []
    while i < len(nums1) and j < len(nums2):
        if nums1[i] <= nums2[j]:
            merged.append(nums1[i])
            i += 1
        else:
            merged.append(nums2[j])
            j += 1
    return merged + nums1[i:] + nums2[j:]

def merge_sort(nums, depth=0):
    print(' '*depth, 'merge_sort', nums)
    if len(nums) < 2:
        return nums
    mid = len(nums) // 2
    return merge(merge_sort(nums[:mid], depth+1),
                 merge_sort(nums[mid:], depth+1),
                 depth+1)
```

## Quick Sort  
To overcome the inefficiencies of space complexity of merge sort, we'll study another divide-and-conquer based algorithm called quick sort.  


```python
def quicksort(nums, start=0, end=None):
    # print('quicksort:', nums, start, end)
    if end is None:
        nums = list(nums)
        end = len(nums) - 1

    if start < end:
        pivot = partition(nums, start, end)
        quicksort(nums, start, pivot-1)
        quicksort(nums, pivot+1, end)
    
    return nums
```

```python
def partition(nums, start=0, end=None):
    # print('partition:', nums, start, end)
    if end is None:
        end = len(nums) - 1 

    # Initializing right and left pointers
    l, r = start, end-1

    # Iterate while they're apart
    while r>l:
        # print(' ', nums, l, r)
        # Increment the left pointer if number is less or equal to pivot
        if nums[l] <= nums[end]:
            l += 1
        
        # Decrement the right pointer if number is greater than pivot
        elif nums[r] > nums[end]:
            r -= 1
        
        # Two out-of-place elements found, swap them
        else:
            nums[l], nums[r] = nums[r], nums[l]

    # print(' ', nums, l, r)
    # Place the pivot between the two parts
    if nums[l] > nums[end]:
        nums[l], nums[end] = nums[end], nums[l]
        return l
    else:
        return end
```
